```python
import random
import numpy as np

class KMeans:
    def __init__(self, n_clusters=2, max_iter=100):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.centroids = None
        self.labels = None
        
    def fit_predict(self, x):
        random_index = random.sample(range(0, x.shape[0]), self.n_clusters)
        self.centroids = x[random_index]
        print(f"Initial centroids:\n{self.centroids}")
        
        for i in range(self.max_iter):
            labels = self.assign_clusters(x)
            new_centroids = self.update_centroids(x, labels)
            
            # Check for convergence (if centroids not changing)
            if np.allclose(self.centroids, new_centroids):
                print(f"Converged after {i+1} iterations.")
                break
            self.centroids = new_centroids
        
        self.labels = labels
        return labels

    def assign_clusters(self, x):
        cluster_group = []
        for row in x:
            distances = []
            for centroid in self.centroids:
                distances.append(np.sqrt(np.dot(row - centroid, row - centroid)))
            cluster_group.append(np.argmin(distances))  # assign closest centroid
        return cluster_group

    def update_centroids(self, x, labels):
        new_centroids = []
        for i in range(self.n_clusters):
            points = x[np.array(labels) == i]
            if len(points) > 0:
                new_centroids.append(points.mean(axis=0))
            else:
                new_centroids.append(self.centroids[i])  # keep old if no points
        return np.array(new_centroids)
```

```python
from sklearn.datasets import make_blobs

# create dummy data
X, y = make_blobs(n_samples=100, centers=3, random_state=42)

# custom KMeans
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X)

print("Final centroids:\n", kmeans.centroids)
print("Predicted labels:\n", labels)
```
